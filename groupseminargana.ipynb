```python
# prompt: reading datset

# Import necessary libraries
import pandas as pd

# Read the dataset from a CSV file
df = pd.read_csv("/content/insurance_claims (1).csv")
```

```python
df.head()
```

```python
# prompt: scatter plot

df.plot.scatter(x="months_as_customer", y="policy_number")
```

```python
# prompt: creative pie chart for less number

import matplotlib.pyplot as plt
# Group the data by the number of months as a customer
grouped_df = df.groupby("months_as_customer").size()

# Create a pie chart of the grouped data
plt.pie(grouped_df, labels=grouped_df.index, autopct="%1.1f%%")
plt.show()
```

```python
# prompt: line chart for less number

import matplotlib.pyplot as plt
less_months = df[df["months_as_customer"] < 10]

plt.plot(less_months["months_as_customer"], less_months["policy_number"])
plt.xlabel("Months as Customer")
plt.ylabel("Policy Number")
plt.title("Line Chart for Customers with Less Than 10 Months")
plt.show()
```
